# **GEOPANDAS**

**GeoPandas is an open source project to make working with geospatial data in python easier.**

**GeoPandas, as the name suggests, extends the popular data science library pandas by adding support for geospatial data.**

**GeoPandas Dependencies:**
**- Pandas:** GeoPandas builds on the Pandas library for data manipulation. Pandas provides data structures like Series and DataFrame, which GeoPandas extends to handle geospatial data.

**- Shapely:** Shapely is used for geometric operations. It provides a set of geometric objects and functions for working with planar geometric objects, such as points, lines, and polygons.

**- Fiona:** Fiona is a Python wrapper around the OGR Simple Features Library. It allows GeoPandas to read and write geospatial vector data formats, such as Shapefiles.

**- Matplotlib:** GeoPandas uses Matplotlib for plotting and visualizing geospatial data. Matplotlib is a popular plotting library in the Python ecosystem.

**- Descartes:** Descartes is used to integrate Matplotlib with Shapely. It helps in visualizing Shapely geometries directly within Matplotlib plots.

**- PyProj:** PyProj is used for coordinate transformations and projections. It provides Python bindings for the PROJ library, which is a library for performing cartographic projections.

# **Type Of Spatial Data**

Geospatial data is created, shared, and stored in many different formats. The two primary data types are raster and vector. Vector data is represented as either points, lines, or polygons

**Vector Data:**

- Representation: Vector data represent geographic features using points, lines, and polygons.
- Elements: The basic elements of vector data are vertices (points), edges (lines), and faces (polygons).
- Geometry Types: Common vector geometry types include Point, LineString, Polygon, MultiPoint, MultiLineString, and MultiPolygon.
- Structure: Vector data are discrete and are characterized by distinct, individual features. Each feature has a well-defined boundary.
- Examples: Examples of vector data include city boundaries, roads, rivers, and administrative boundaries.

**Raster Data:**

- Representation: Raster data represent geographic features using a grid of cells or pixels.
- Elements: The basic elements of raster data are cells or pixels, each having a value representing a specific attribute.
- Geometry Types: There is no concept of points, lines, or polygons in raster data. Instead, the entire dataset is a regular grid.
- Structure: Raster data are continuous and represent a surface or field. Each cell has a value, and neighboring cells share boundaries.
- Examples: Examples of raster data include satellite imagery, elevation models, and land cover maps.

**Geospatial Data Formats**
**Vector Formats:**

**- SHP: Shapefile**

The ESRI Shapefile has become an industry standard geospatial data format, and is compatible to some extent with practically all recently released GIS software. To have a complete shapefile, you must have at least 3 files with the same prefix name and with the following extensions: .shp = shapefile, .shx = header and .dbf = associated database file. Additionally, you may have a .prj = Projection file, a .lyr = layer file, and other index files. All these files must be saved in the same workspace.

**- SDC: Smart Data Compression**

SDC is ESRI's highly compressed format, which is directly readable by ArcGIS software.

**- GDB: Geodatabase**

The file geodatabase is a collection of geographic datasets of various types, with the most basic types being vector, raster, and tabular data. There are three types of geodatabases: file, personal, and ArcSDE. Geodatabases are the native data format for ESRI's ArcGIS.

**- ArcInfo Coverage**

An ArcInfo coverage has been phased out and is rarely seen/used today. It has largely been replaced by the geodatabase format.  Coverages do not have an individual file extension. Instead it is composed of two folders within a "workspace" which each contain multiple files. One of the two folders carries the name of the coverage, and contains a number of various .adf files. The other folder is an "info" folder, which typically contains .dat and .nit files for all the coverages and grids in the workspace.

**- E00: Arc Export or Interchange Format**

.e00 files are also rarely seen/used today, but are essentially ArcInfo Interchange or export files, used to conveniently copy and move ArcInfo GIS coverages (see above) and grids (see below). An .e00 file must be "imported" and converted in order to use the data in ArcGIS or other GIS software.

**Raster Formats:**

**- ArcInfo Grid**

An ArcInfo Grid does not have an individual file extension. Instead it is composed of two folders within a "workspace" which each contain multiple files. One of the two folders carries the name of the grid, and contains a number of various .adf files. The other folder is an "info" folder, which typically contains .dat and .nit files for all the coverages and grids in the workspace.

**- Band Interleaved by Line (BIL), Band Interleaved by Pixel (BIP), and Band Sequential (BSQ).**

BIL, BIP, and BSQ are formats produced by remote-sensing systems. The primary difference among them is the technique used to store brightness values captured simultaneously in each of several colors or spectral bands.

**- DEM (Digital Elevation Model)**

DEM is a raster format used by the USGS to record elevation information. Unlike other raster file formats, DEM cells do not represent color brightness values, but rather the elevations of points on the earth’s surface.

**- GeoTIFF**

As part of the header of the TIFF file, this provides the Lat/Long boundary extent of the data.

**- LiDAR**

Light Detection and Ranging (LiDAR), is a remote sensing method that uses light in the form of a pulsed laser to measure ranges (variable distances) to the Earth. File formats vary by deliverable - but raw LiDAR point cloud data has a .LAS file extension. A DEM can be one deliverable from LiDAR.


# **Coordinate Reference System (CRS) in Geopandas: A Brief Introduction**

**What is a Coordinate Reference System (CRS)?**

A Coordinate Reference System (CRS) is a system that uses coordinates to uniquely determine the position of a point or geometric element on the Earth's surface. It provides a framework for mapping locations, enabling accurate spatial representation and analysis.

# **Why Does CRS Matter in Geopandas?**

**Spatial Accuracy:**

Different CRSs represent the Earth's surface differently. Choosing the right CRS is crucial for accurate spatial analysis.
The Earth is not flat, so a flat representation introduces distortions. CRS helps manage these distortions.

**Spatial Analysis:**

When performing spatial operations or analyses, datasets must be in the same CRS. Mismatched CRSs can lead to inaccurate results.
CRS ensures that geometric operations (e.g., intersections, buffers) are meaningful.

**Visualization:**

Proper CRS selection enhances map visualization. Maps are more meaningful when features are accurately positioned on the Earth's surface.

**Data Integration:**

Datasets from various sources may use different CRSs. Transforming them to a common CRS facilitates integration and analysis.

**Distance and Area Calculations:**

Calculations involving distances or areas depend on the CRS. Using an inappropriate CRS can lead to incorrect measurements.

**Common CRS Properties:**

**EPSG Code:** A unique identifier for a CRS defined by the European Petroleum Survey Group (EPSG).

Units: Determines the unit of measurement (e.g., meters, degrees).

Projection: Describes how the 3D Earth is projected onto a 2D surface.

Datum: Specifies the reference point for measurements.

GeoPandas inherits a number of useful methods and attributes from the shapely package. We’re going to cover four of them.

```python
districts['geometry']
```

**Area**

```python
districts['geometry']
```

```python
# Area in km2

districts['area(km^2)'] = districts['geometry'].area/1000000
```

```python
districts.head()
```

**Bounds**

The .bounds attribute returns a DataFrame with columns minx, miny, maxx, and maxy, representing the bounding box of each geometry. This information is useful for tasks such as setting map extents or spatial indexing.

```python
districts[['minx','miny','maxx','maxy']] = districts['geometry'].bounds
```

```python
districts.head()
```

**Centroid**

The .centroid attribute returns a GeoSeries of the centroids of each geometry. Centroids are often used as representative points for regions or polygons.

```python
districts['centroid'] = districts['geometry'].centroid
```

```python
districts.head()
```

**Distance**

Say we want to calculate the distance from the famous Sagrada Familia church, located in the Eixample district, to the centroids of every district in Barcelona, and then add the distances (in kilometers) in a new column.

```python
# importing a point class to help us create a point object
from shapely.geometry import Point

# creating a point object with the cordinates of the Sagrada Church
sagrada_fam = Point(2.1743680500855005, 41.403656946781304)
sagrada_fam = gpd.GeoSeries(sagrada_fam, crs=4326)

# converting the crs
sagrada_fam= sagrada_fam.to_crs(epsg=2062)
```

```python
districts.head()
```

```python
districts.apply(lambda row: float(row['centroid'].distance(sagrada_fam))/1000,axis=1)
```

```python
districts['km from sagrada'] = districts.apply(lambda row: float(row['centroid'].distance(sagrada_fam))/1000,axis=1)
```
