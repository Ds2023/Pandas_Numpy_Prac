Turning item price into a float

```python
chipo['item_price']=chipo['item_price'].str.replace('$','').astype(float)
```

Revenue generated for the period

```python
revenue = (chipo['quantity'] * chipo['item_price']).sum()
revenue
```

Orders made in the period

```python
tot_orders = chipo['order_id'].count()
tot_orders
```

Average revenue amount per order

```python
aver_rev = revenue/tot_orders
aver_rev
```
