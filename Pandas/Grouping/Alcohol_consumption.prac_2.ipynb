```python
drinks = pd.read_csv('drinks.csv')
drinks.head()
```

Continent drinking more beer on average

```python
drinks.groupby('continent')['beer_servings'].aggregate('max')
```

Wine consumption statistics per continent

```python
drinks.groupby('continent')['wine_servings'].describe()
```

Mean alcohol consumption per continent

```python
drinks.groupby('continent').mean()
```

Median alcohol consumption per continent

```python
drinks.groupby('continent').median()
```

Mean,min and max values for spirit consumption

```python
drinks.groupby('country')['spirit_servings'].aggregate(['min', 'mean', 'max'])
```
